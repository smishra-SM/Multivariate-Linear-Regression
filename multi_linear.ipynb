```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
from sklearn import linear_model
%matplotlib inline
```

```python
df = pd.read_csv("Real estate.csv")
df.head(7)
```

```python
cdf = df[['house_age','distance_to_nearest_MRT_station','number_of_convenience_stores','house_price']]
cdf.head(7)
```

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['house_age','distance_to_nearest_MRT_station','number_of_convenience_stores']])
train_y = np.asanyarray(train[['house_price']])
regr.fit (train_x, train_y)
print ('Coefficients: ', regr.coef_)
```

```python
y_hat= regr.predict(test[['house_age','distance_to_nearest_MRT_station','number_of_convenience_stores']])
test_x = np.asanyarray(test[['house_age','distance_to_nearest_MRT_station','number_of_convenience_stores']])
test_y = np.asanyarray(test[['house_price']])
print("Mean square error MSE: %.2f"
      % np.mean((y_hat - test_y) ** 2))

```
